```python
#引進套件
import matplotlib.pyplot as plt
import matplotlib.style
import pandas as pd
import numpy as np

#在畫圖時使用 ggplot 的風格
matplotlib.style.use('ggplot')
```

```python
#讀取 csv
df = pd.read_csv('http://bit.ly/kaggletrain')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth',100)
```

```python
#利用 subplots() 在一張畫布上放四張圖，並且調整圖的大小
fig, axes = plt.subplots(2,2,figsize = (10, 10))

#第一題

#從原先表格中個別篩選出性別為男和性別為女的資料，並且使用 Python 中的 len() 來獲取資料的長度，藉此了解男生和女生的具體人數
maleNumber = len(df[df["Sex"] == "male"])
femaleNumber = len(df[df['Sex'] == "female"])

#設定長條圖的 X 軸和 Y 軸
x = ["male", "female"]
y = [maleNumber, femaleNumber]

#畫出符合題目的長條圖
axes[0][0].bar(x,y)

#第二題

#從原先的表格中篩選出乘客的年齡資料
x = df["Age"]

# 畫出符合題目的直方圖
axes[0][1].hist(x, bins = 20)

#第三題

#先使用 query() 篩選需要的資料
liveOne = len(df.query('Pclass == 1 and Survived == 1'))
deadOne = len(df.query('Pclass == 1 and Survived == 0'))

liveTwo = len(df.query('Pclass == 2 and Survived == 1'))
deadTwo = len(df.query('Pclass == 2 and Survived == 0'))

liveThree = len(df.query('Pclass == 3 and Survived == 1'))
deadThree = len(df.query('Pclass == 3 and Survived == 0'))

#先把 X 軸要的標籤寫出來
label = ["Pclass1", "Pclass2", "Pclass3"]

#設定長條圖的寬度
width = 0.4

#設定長條圖的 X 軸和 Y 軸，並且利用 Python 提供的產生器方式來錯開 X1 和 X2 的距離
x1 = [1,2,3]
x2 = [num + width for num in x1]
y1 = [deadOne, deadTwo, deadThree]
y2 = [liveOne, liveTwo, liveThree]

#畫出符合題目的長條圖
axes[1][0].bar(x1, y1, tick_label = label, width = width, label = "Dead")
axes[1][0].bar(x2, y2, width = width, label = "Live")

#調整圖標的位置
axes[1][0].legend()

#第四題（正式）

#篩選題目需要用到的欄位
needDf = df[["Age", "Fare", "Survived"]]

#分別篩選生還或未生還的人數
survived = needDf[needDf["Survived"] == 1]
dead = needDf[needDf["Survived"] == 0]

#設定散佈圖的 X 軸和 Y 軸 
x1 = dead["Age"]
x2 = survived["Age"]
y1 = dead["Fare"]
y2 = survived["Fare"]

#畫出符合題目的散佈圖，並且更換點的圖示
axes[1][1].scatter(x1, y1, marker = "x")
axes[1][1].scatter(x2, y2, marker = "s")

plt.show()
```
